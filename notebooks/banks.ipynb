```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data
dataframes = {
   ' Boa_data' :pd.read_excel("../data/BANKS AD DATA.xlsx",sheet_name="BOA"),
    'Cbe_data': pd.read_excel("../data/BANKS AD DATA.xlsx",sheet_name="CBE"),
    'GlobalBank_data'  : pd.read_excel("../data/BANKS AD DATA.xlsx",sheet_name="Global Bank"),
    'Wegagen_data':  pd.read_excel("../data/BANKS AD DATA.xlsx",sheet_name="Wegagen")
}
```

```python
def read_data(file_path):
    return pd.read_excel(file_path)
```

```python
def calculate_total_views(dataframes):
    total_views = {name: df['View'].sum() for name, df in dataframes.items()}
    return pd.DataFrame.from_dict(total_views, orient='index', columns=['Total Views'])
```

```python

# Analysis 1: Compare the total number of views across the datasets
total_views_df = calculate_total_views(dataframes)

print(total_views_df)
```

```python
def calculate_avg_views_per_post(dataframes):
    avg_views_per_post = {name: df['View'].mean() for name, df in dataframes.items()}
    print("\nAverage Views per Post:")
    return pd.DataFrame.from_dict(avg_views_per_post, orient='index', columns=['Average Views per Post'])
```

```python
# Analysis 2: Compare the average views per post for each dataset
avg_views_df = calculate_avg_views_per_post(dataframes)
print("\nAverage Views per Post:")
print(avg_views_df)

```

```python
def count_posts_by_time_of_day(dataframes):
    time_of_day_counts = {name: df['Time of day'].value_counts() for name, df in dataframes.items()}
    return time_of_day_counts
```

```python
# Analysis 3: Compare the number of posts made during different times of the day
time_of_day_counts = count_posts_by_time_of_day(dataframes)
print("\nNumber of Posts by Time of Day:")
for name, counts in time_of_day_counts.items():
    print(f"{name}:\n{counts}")
```

```python
def count_posts_by_day_of_week(dataframes):
    day_of_week_counts = {name: df['Date'].dt.day_name().value_counts() for name, df in dataframes.items()}
    return day_of_week_counts
def count_posts_by_day_of_week(dataframes):
    day_of_week_counts = {}
    for name, df in dataframes.items():
        if 'Date' in df.columns and pd.api.types.is_datetime64_any_dtype(df['Date']):
            day_of_week_counts[name] = df['Date'].dt.day_name().value_counts()
        else:
            day_of_week_counts[name] = "Date column not found or not in datetime format"
    return day_of_week_counts
```

```python
# Analysis 4: Compare the number of posts made on different days of the week
day_of_week_counts = count_posts_by_day_of_week(dataframes)
print("\nNumber of Posts by Day of Week:")
for name, counts in day_of_week_counts.items():
    print(f"{name}:\n{counts}")
```

```python
def plot_total_views_bar_chart(df):
    df.plot(kind='bar', legend=False)
    plt.xlabel('Bank')
    plt.ylabel('Total Views')
    plt.title('Total Views by Bank')
    plt.show()
```

```python

plot_total_views_bar_chart(total_views_df)
```

```python
def plot_avg_views_bar_chart(df):
    df.plot(kind='bar', legend=False)
    plt.xlabel('Bank')
    plt.ylabel('Average Views per Post')
    plt.title('Average Views per Post by Bank')
    plt.show()
```

```python
plot_avg_views_bar_chart(avg_views_df)
```

```python

def plot_posts_by_time_of_day_pie_charts(dataframes):
    for name, counts in dataframes.items():
        counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
        plt.title(f"Posts by Time of Day - {name}")
        plt.ylabel('')
        plt.show()
```

```python
plot_posts_by_time_of_day_pie_charts(time_of_day_counts)
```
